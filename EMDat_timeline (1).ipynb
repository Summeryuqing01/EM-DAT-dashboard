```python
data = pd.read_excel("/content/drive/MyDrive/A_emdat/public_emdat_custom_request_2025-01-13_adee2b78-5358-40ab-9ea2-12c160f21f8c.xlsx")
```

```python
np.unique(data['Disaster Subgroup'])
```

```python
new_data = data
new_data['Start Month'] = new_data['Start Month'].fillna(1)
new_data['Start Day'] = new_data['Start Day'].fillna(1)
```

```python
new_data['End Month']= new_data['End Month'].fillna(12)
new_data['End Day'] = new_data['End Day'].fillna(28)
```

```python
dateInt= new_data['Start Year'].astype(str) + new_data['Start Month'].astype(int).astype(str).str.zfill(2)+ new_data['Start Day'].astype(int).astype(str).str.zfill(2)
new_data['Start Date'] = pd.to_datetime(dateInt)
```

```python
dateInt= new_data['End Year'].astype(str) + new_data['End Month'].astype(int).astype(str).str.zfill(2)+ new_data['End Day'].astype(int).astype(str).str.zfill(2)
new_data['End Date'] = pd.to_datetime(dateInt)
```

```python
display_timeline(
   data=new_data,
   title="Asia Disaster from 2000 by subregion",
   time_column="Start Date",
   group_by="Subregion",
   source_columns=["Disaster Group", "Disaster Subtype", "Country",'Subregion','Location'],
   legend="left",
   height=300,
)
```

```python
sea = new_data[new_data['Subregion']=='South-eastern Asia']
display_timeline(
   data=sea,
   title="South East Asia Disaster from 2000 by country",
   time_column="Start Date",
   group_by="Country",
   source_columns=["Disaster Group", "Disaster Subtype", "Country",'Subregion','Location'],
   legend="left",
   height=300,
)
```

```python
import plotly.express as px
fig = px.timeline(sea, x_start="Start Date", x_end="End Date", y='Country',
                  color="Disaster Subgroup",
                  opacity = 0.6,
                  hover_data = ['Start Date','End Date',"Disaster Group", "Disaster Subtype", "Country",'Subregion','Location'])
fig.update_yaxes(autorange="reversed")
fig.show(config={'scrollZoom': True})
```

Note:
* The Start Day, Start Month, and Start Year are combined to create the new time column. There are some missing months and days for mostly large scale disasters that are hard to measure, those values are set to 1.
* Similar to the start time, End Day and Month are missing for disasters such as viral outbreak or drought. For simplicity, missing end months are set to 12 and missing end days are set to 28
