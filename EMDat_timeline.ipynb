```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

#!pip install msticpy
from msticpy.vis import mp_pandas_plot
from msticpy.vis.timeline import display_timeline
```

```python
pd.set_option("display.max_columns", None)
```

```python
data = pd.read_excel("/content/public_emdat_custom_request_2025-01-13_adee2b78-5358-40ab-9ea2-12c160f21f8c.xlsx")
```

```python
data.info()
```

```python
new_data = data
new_data['Start Month'] = new_data['Start Month'].fillna(1)
new_data['Start Day'] = new_data['Start Day'].fillna(1)
```

```python
dateInt= new_data['Start Year'].astype(str) + new_data['Start Month'].astype(int).astype(str).str.zfill(2)+ new_data['Start Day'].astype(int).astype(str).str.zfill(2)
new_data['Date'] = pd.to_datetime(dateInt)
```

```python
display_timeline(
   data=new_data,
   title="Asia Disaster from 2000 by subregion",
   time_column="Date",
   group_by="Subregion",
   source_columns=["Disaster Group", "Disaster Subtype", "Country",'Subregion','Location'],
   legend="left",
   height=300,
)
```

```python
sea = new_data[new_data['Subregion']=='South-eastern Asia']
display_timeline(
   data=sea,
   title="South East Asia Disaster from 2000 by country",
   time_column="Date",
   group_by="Country",
   source_columns=["Disaster Group", "Disaster Subtype", "Country",'Subregion','Location'],
   legend="left",
   height=300,
)
```

Note:
* The Start Day, Start Month, and Start Year are combined to create the new time column. There are some missing months and days for mostly large scale disasters that are hard to measure, those values are set to 1.
* The timeline is plotted with the start date, it doesn't reflect the duration of the event.
